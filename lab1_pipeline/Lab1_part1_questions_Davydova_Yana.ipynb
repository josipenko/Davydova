## ex. 1

$$  
y = x^Tx,  \quad x \in \mathbb{R}^N
$$

$$
\frac{dy}{dx} = 2 x.
$$

**Solution**.
First of all, we notice that $y$ is a scalar. Indeed,
$$
y = \sum_{j = 1}^{N} x_{j}^{2}.
$$
The derivative $dy/dx$ is a column vector. The $i$-th component of this vector is
$$
\left(\frac{dy}{dx}\right)_{i} = \frac{\partial}{\partial x_{i}}\left( \sum_{j = 1}^{N} x_{j}^{2}\right) = \sum_{j = 1}^{N} \frac{\partial}{\partial x_{i}} x_{j}^{2}.
$$
The derivative in the sum is nonzero only if $i = j$. This can be expressed using the Kronecker delta symbol
$$
\frac{\partial}{\partial x_{i}} x_{j}^{2} = \delta_{ij} \, 2 x_{j}.
$$
So,
$$
\left(\frac{dy}{dx}\right)_{i} = 2 \sum_{j = 1}^{N} \delta_{ij} x_{j} = 2 x_{i},
$$
or in matrix notation
$$
\frac{dy}{dx} = 2 x.
$$

## ex. 2

$$ y = tr(AB) \quad A,B \in \mathbb{R}^{N \times N} $$

$$
\frac{dy}{dA} = B^{T}.
$$

**Solution**.  First of all, we notice that $y$ is a scalar. Indeed,
$$
(AB)_{in} = \sum_{j=1}^{N} a_{ij} b_{jn},
$$
so
$$
y = \sum_{i=1}^{N} \sum_{j=1}^{N} a_{ij} b_{ji}.
$$
The derivative $dy/dA$ is a matrix in $\mathbb{R}^{N \times N}$. The component of this vector with indices $m$ and $n$ is
$$
\left(\frac{dy}{dA}\right)_{mn} = \frac{\partial}{\partial a_{mn}} \sum_{i=1}^{N} \sum_{j=1}^{N} a_{ij} b_{ji} = \sum_{i=1}^{N} \sum_{j=1}^{N}  \frac{\partial a_{ij}}{\partial a_{mn}} b_{ji}.
$$
Taking into account that
$$
\frac{\partial a_{ij}}{\partial a_{mn}} = \delta_{im}\delta_{jn},
$$
we have
$$
\left(\frac{dy}{dA}\right)_{mn} = \sum_{i=1}^{N} \sum_{j=1}^{N}  \delta_{im}\delta_{jn} b_{ji} = b_{nm},
$$
or in matrix form (we notice that $mn$-component in the left-hand side corresponds to $nm$-component in the right-hand side, so the transposition is necessary)
$$
\frac{dy}{dA} = B^{T}.
$$

## ex. 3

$$  
y = x^TAc , \quad A\in \mathbb{R}^{N \times N}, x\in \mathbb{R}^{N}, c\in \mathbb{R}^{N}
$$

$$
\frac{dy}{dx} = A c,
$$

$$
\frac{dy}{dA} = x c^{T}.
$$

**Solution**. First of all, we notice that $y$ is a scalar. Indeed,
$$
y = \sum_{i = 1}^{N} \sum_{j = 1}^{N} x_{i} A_{ij} c_{j}.
$$
The derivative $dy/dx$ is a column vector. The $n$-th component of this vector is
$$
\left(\frac{dy}{dx}\right)_{n} = \sum_{i = 1}^{N} \sum_{j = 1}^{N} \frac{\partial x_{i}}{\partial x_{n}} A_{ij} c_{j} = \sum_{i = 1}^{N} \sum_{j = 1}^{N} \delta_{in} A_{ij} c_{j} = \sum_{j = 1}^{N} A_{nj} c_{j},
$$
or in matrix notation
$$
\frac{dy}{dx} = A c.
$$

Here we use the trace cyclic property and the fact that $y$ is a scalar, so we can formally write trace operation in front of the expression.
$$
\frac{dy}{dA} = \frac{d}{dA} \left( x^{T} A c \right) =  \frac{d}{dA} tr\left(x^{T} A c \right) = \frac{d}{dA} tr\left(A c x^{T} \right) = |\text{ex.2}| = \left(c x^{T}\right)^{T} = x c^{T}.
$$

## ex. 4

Classic matrix factorization example. Given matrix $X$ you need to find $A$, $S$ to approximate $X$. This can be done by simple gradient descent iteratively alternating $A$ and $S$ updates.
$$
J = || X - AS ||_F^2  , \quad A\in \mathbb{R}^{N \times R} , \quad S\in \mathbb{R}^{R \times M}
$$
$$
\frac{dJ}{dS} = ?
$$

You may use one of the following approaches:

#### First approach
Using ex.2 and the fact:
$$
|| X ||_F^2 = tr(XX^T)
$$
it is easy to derive gradients (you can find it in one of the refs).

#### Second approach
You can use *slightly different techniques* if they suits you. Take a look at this derivation:
<img src="https://github.com/girafe-ai/ml-course/blob/22f_basic/homeworks/lab01_ml_pipeline/grad.png?raw=1">
(excerpt from [Handbook of blind source separation, Jutten, page 517](https://books.google.ru/books?id=PTbj03bYH6kC&printsec=frontcover&dq=Handbook+of+Blind+Source+Separation&hl=en&sa=X&ved=0ahUKEwi-q_apiJDLAhULvXIKHVXJDWcQ6AEIHDAA#v=onepage&q=Handbook%20of%20Blind%20Source%20Separation&f=false), open for better picture).

#### Third approach
And finally we can use chain rule!
let $ F = AS $

**Find**
$$
\frac{dJ}{dF} =  2(A S - X) \in \mathbb{R}^{N \times M}
$$
and
$$
\frac{dF}{dS} = I \otimes A, \quad I \in \mathbb{R}^{M \times M} \ \text{identity matrix}
$$
(the shape should be $ NM \times RM )$.

Now it is easy to get desired gradients:
$$
\frac{dJ}{dS} =  2 A^{T}\left( A S - X\right).
$$

**Solution**.
Chain rule in the component notation has the form
$$
\frac{\partial J}{\partial s_{rm}} = \sum_{i=1}^{N} \sum_{j=1}^{M} \frac{\partial J}{\partial f_{ij}} \frac{\partial f_{ij}}{\partial s_{rm}},
$$
or in matrix notation (see ex.2)
$$
\frac{\partial J}{\partial s_{rm}} = tr \left( \left(\frac{d J}{d F}\right)^{T} \frac{\partial F}{\partial s_{rm}} \right).
$$
$$
\frac{dJ}{d F}= \frac{d}{d F}\left(|| X - F ||_F^2\right) = \frac{d}{d F} tr\left[\left(X - F\right)^{T} \left(X - F\right)\right] = |Y := X - F, dF = -dY| = -\frac{d}{d Y} tr(Y^{T}Y) = |\text{ex.2}| = -2 Y = 2(F - X) = 2(A S - X).
$$
We define $dF/dS$ as follows
$$
\frac{d F}{d S} := \begin{pmatrix}
\frac{\partial f_{11}}{\partial s_{11}} & \frac{\partial f_{11}}{\partial s_{21}} & \ldots & \frac{\partial f_{11}}{\partial s_{RM}} \\
\frac{\partial f_{21}}{\partial s_{11}} & \frac{\partial f_{21}}{\partial s_{21}} & \ldots & \frac{\partial f_{21}}{\partial s_{RM}} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_{N1}}{\partial s_{11}} & \frac{\partial f_{N1}}{\partial s_{21}} & \ldots & \frac{\partial f_{N1}}{\partial s_{RM}} \\
\frac{\partial f_{12}}{\partial s_{11}} & \frac{\partial f_{12}}{\partial s_{21}} & \ldots & \frac{\partial f_{12}}{\partial s_{RM}}\\
\frac{\partial f_{22}}{\partial s_{11}} & \frac{\partial f_{22}}{\partial s_{21}} & \ldots & \frac{\partial f_{22}}{\partial s_{RM}}\\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_{NM}}{\partial s_{11}} & \frac{\partial f_{NM}}{\partial s_{21}} & \ldots & \frac{\partial f_{NM}}{\partial s_{RM}}
\end{pmatrix} \quad \in \mathbb{R}^{NM \times RM}.
$$
Component of this matrix are
$$
\frac{\partial f_{ij}}{\partial s_{rm}} =  \frac{\partial}{\partial s_{rm}}\left(\sum_{k=1}^{N}a_{ik} s_{kj}\right) = \sum_{k=1}^{N}a_{ik} \frac{\partial s_{kj}}{\partial s_{rm}} = \sum_{k=1}^{N} a_{ik} \delta_{kr} \delta_{jm} = a_{ir} \delta_{jm},
$$
or in the matrix form
$$
\frac{d F}{d S} = I \otimes A,
$$
where $I$ is a $M \times M$ identity matrix.

Finally,
$$
\frac{\partial J}{\partial s_{rm}} = 2 \sum_{i=1}^{N} \sum_{j=1}^{M} \left( A S - X\right)_{ij}a_{ir}\delta_{jm} = 2 \sum_{i=1}^{N} \left( A S - X\right)_{im}a_{ir}= 2 \sum_{i=1}^{N} \left(A^{T}\right)_{ri}\left( A S - X\right)_{im} = 2 A^{T}\left( A S - X\right).
$$

### Question 2

We can also use other distance metrics such as L1 distance.
For pixel values $p_{ij}^{(k)}$ at location $(i,j)$ of some image $I_k$,

the mean $\mu$ across all pixels over all images is $$\mu=\frac{1}{nhw}\sum_{k=1}^n\sum_{i=1}^{h}\sum_{j=1}^{w}p_{ij}^{(k)}$$
And the pixel-wise mean $\mu_{ij}$ across all images is
$$\mu_{ij}=\frac{1}{n}\sum_{k=1}^np_{ij}^{(k)}.$$
The general standard deviation $\sigma$ and pixel-wise standard deviation $\sigma_{ij}$ is defined similarly.

Which of the following preprocessing steps will not change the performance of a Nearest Neighbor classifier that uses L1 distance? Select all that apply.
1. Subtracting the mean $\mu$ ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu$.)
2. Subtracting the per pixel mean $\mu_{ij}$  ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu_{ij}$.)
3. Subtracting the mean $\mu$ and dividing by the standard deviation $\sigma$.
4. Subtracting the pixel-wise mean $\mu_{ij}$ and dividing by the pixel-wise standard deviation $\sigma_{ij}$.
5. Rotating the coordinate axes of the data.

*Your Answer:*
**1, 2, 3**

*Your Explanation:*
1. $$
|| \tilde{x}_{\mathrm{test}} - \tilde{x}_{\mathrm{train}}^{(k)}||_{1} = || \left(\tilde{x}_{\mathrm{test}} - \mu \right) - \left(\tilde{x}_{\mathrm{train}}^{(k)} - \mu \right)||_{1} =  || x_{\mathrm{test}} - x_{\mathrm{train}}^{(k)}||_{1},
$$
where, $x_{\mathrm{test}}$ is a test image vector and $x_{\mathrm{train}}^{(k)}$ is a $k$-th training image vector.
2. $$
|| \tilde{x}_{\mathrm{test}} - \tilde{x}_{\mathrm{train}}^{(k)}||_{1} = \sum_{i=1}^{h}\sum_{j=1}^{w} \left|\left(\tilde{p}_{\mathrm{test}}\right)_{ij} - \left(\tilde{p}_{\mathrm{train}}^{(k)}\right)_{ij}\right| =  \sum_{i=1}^{h}\sum_{j=1}^{w} \left|\left(p_{\mathrm{test}}\right)_{ij} - \mu_{ij} - \left(\left(p_{\mathrm{train}}^{(k)}\right)_{ij} - \mu_{ij}\right)\right| = \sum_{i=1}^{h}\sum_{j=1}^{w} \left|\left(p_{\mathrm{test}}\right)_{ij} - \left(p_{\mathrm{train}}^{(k)}\right)_{ij}\right|= || x_{\mathrm{test}} - x_{\mathrm{train}}^{(k)}||_{1}.
$$
3. $$
|| \tilde{x}_{\mathrm{test}} - \tilde{x}_{\mathrm{train}}^{(k)}||_{1} =  \sum_{i=1}^{h}\sum_{j=1}^{w} \left|\left(\tilde{p}_{\mathrm{train}}\right)_{ij} - \left(\tilde{p}_{\mathrm{test}}^{(k)}\right)_{ij}\right| =  \sum_{i=1}^{h}\sum_{j=1}^{w} \left|\frac{\left(p_{\mathrm{test}}\right)_{ij} - \mu}{\sigma} - \frac{\left(p_{\mathrm{test}}^{(k)}\right)_{ij} - \mu}{\sigma}\right| =\frac{1}{\sigma} \sum_{i=1}^{h}\sum_{j=1}^{w} \left|\left(p_{\mathrm{test}}\right)_{ij} - \left(p_{\mathrm{test}}^{(k)}\right)_{ij}\right|= \frac{1}{\sigma} || x_{\mathrm{test}} - x_{\mathrm{train}}^{(k)}||_{1}.
$$
Thus, the L1 distance is effectively just being scaled by a fixed factor. This would cause the nearest neighbors of a point to remain the same and thus would not affect the performance of the classifier. Performing feature scaling does help facilitate learning by bounding the gradients and thus making gradient descent converge faster and speed up training, but this does not affect classifier accuracy.
4. $$
|| \tilde{x}_{\mathrm{test}} - \tilde{x}_{\mathrm{train}}^{(k)}||_{1} =  \sum_{i=1}^{h}\sum_{j=1}^{w} \left|\left(\tilde{p}_{\mathrm{train}}\right)_{ij} - \left(\tilde{p}_{\mathrm{test}}^{(k)}\right)_{ij}\right| =  \sum_{i=1}^{h}\sum_{j=1}^{w} \left|\frac{\left(p_{\mathrm{test}}\right)_{ij} - \mu_{ij}}{\sigma_{ij}} - \frac{\left(p_{\mathrm{test}}^{(k)}\right)_{ij} - \mu_{ij}}{\sigma_{ij}}\right| = \sum_{i=1}^{h}\sum_{j=1}^{w} \frac{1}{\sigma_{ij}}\left|\left(p_{\mathrm{test}}\right)_{ij} - \left(p_{\mathrm{test}}^{(k)}\right)_{ij}\right|.
$$
Thus, we are scaling different dimensions differently. The nearest neighbors could thus change and this would in turn affect the performance of the classifier.
5. L1 distance is not invariant to the rotation of the coordinate axes, unlike L2. The distance between points changes with the rotation of axes and thus, the performance of the classifier can potentially be affected.
Lets consider one train vector $y = (1, 0)^{T}$ and two test vectors $x_{1} = (0, 1)^{T}$ and $x_{2} = (1, 2)^{T}$.
Initially,
$$
||x_{1} - y|| = |0 - 1| + |1 - 0| = 2, \\
||x_{2} - y|| = |1 - 1| + |2 - 0| = 2.
$$
Thus, both $x_{1}$ and $x_{2}$ are at the same distance from $y$.
Next, we consider $\pi/4$ rotation matrix
$$
A = \begin{pmatrix}
\cos\left(\frac{\pi}{4}\right) & -\sin\left(\frac{\pi}{4}\right) \\
\sin\left(\frac{\pi}{4}\right) & \cos\left(\frac{\pi}{4}\right)
\end{pmatrix} = \begin{pmatrix}
\frac{1}{\sqrt{2}} & - \frac{1}{\sqrt{2}}\\
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}
\end{pmatrix} = \frac{1}{\sqrt{2}}\begin{pmatrix}
1 & -1\\
1 & 1
\end{pmatrix}
$$.
So,
$$
\tilde{x}_{1} = A x_{1} = \frac{1}{\sqrt{2}}\begin{pmatrix}
-1 \\
1
\end{pmatrix},
$$
$$
\tilde{x}_{2} = A x_{2} = \frac{1}{\sqrt{2}}\begin{pmatrix}
-1 \\
3
\end{pmatrix},
$$
$$
\tilde{y} = A y = \frac{1}{\sqrt{2}}\begin{pmatrix}
1 \\
1
\end{pmatrix}.
$$
Thus,
$$
||\tilde{x}_{1} - \tilde{y}|| = \frac{1}{\sqrt{2}}\left(|- 1 - 1| + |1 - 1\right) = \frac{2}{\sqrt{2}}, \\
||\tilde{x}_{2} - \tilde{y}|| = \frac{1}{\sqrt{2}}\left(|-1 - 1| + | 3 - 1|\right) = \frac{4}{\sqrt{2}}.
$$
Thus,
$$
||\tilde{x}_{1} - \tilde{y}|| < ||\tilde{x}_{2} - \tilde{y}||.
$$



### Question 3

Which of the following statements about $k$-Nearest Neighbor ($k$-NN) are true in a classification setting, and for all $k$? Select all that apply.
1. The decision boundary (hyperplane between classes in feature space) of the k-NN classifier is linear.
2. The training error of a 1-NN will always be lower than that of 5-NN.
3. The test error of a 1-NN will always be lower than that of a 5-NN.
4. The time needed to classify a test example with the k-NN classifier grows with the size of the training set.
5. None of the above.

*Your Answer:*
**2, 4**

*Your Explanation:*
1. **False**. The decision boundary of the k-NN classifier is not linear. If you consider a dataset where the classes belong to concentric circles, the decision boundaries in this case will follow the curvature of the concentric circles.

2. **True**. The training error of a 1-NN will always be lower than that of 5-NN because for each training example, its nearest neighbor is always going to be itself, i.e., error of 1-NN will be zero.

3. **False**. The test error of a 1-NN will not always be lower than 5-NN. Lets consider an example. Suppose $x_{\mathrm{train}} = (1, 2, 3, 4, 5)^{T}$ and $y_{\mathrm{train}} = (0, 0, 0, 0, 1)^{T}$. For a test sample $x = 6$ with $y = 0$,
$y_{\textrm{pred}}$ would be $1$ for 1-NN (error = 100%) while
$y_{\textrm{pred}} = 0$ for 5-NN (thus, error = 0%).

4. **True**. The testing phase of k-NN is essentially performing comparisons of each test sample with the entire training set, which needs one full pass through the training set. The training phase of k-NN, which consists of remembering the traning set would also grow with the size of the training set.
