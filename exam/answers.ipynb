__1.__ __Методами машинного обучения (не статистическими тестами) показать, что разбиение на трейн и тест репрезентативно.__

Под репрезентативностью понимается схожесть тренировочных и тестовых данных, т.е. мы не сможем отличить является ли единичный экземпляр данных тренировочным или тестовым. Проверим это с помощью алгоритма машинного обучения под названием случайный лес.
1. Берём тренировочные и тестовые данные и убираем из них столбец с целевыми значениями. В качестве целевых значений будем использовать индикаторы того, что строка принадлежит тренировочным данным (т.е. тренировочные данные имеют индикатор 1 в поле целевых значений, а тестовые -- 0).
2. Объединяем модифицированные в пункте 1 тренировочные и тестовые данные.
3. Разделяем объединенные данные на матрицу признаков и целевой вектор.
4. Используем алгоритм случайного дерева с кросс-валидацией c параметром, позволяющим оценить качество бинарной классификации (ROC_AUC).
5. Если величина ROC_AUC равна 0.5, то случайный лес не видит разницы, к каким данным может принадлежат строка к тренировочным или тестовым. Это показывает, что разбиение на тестовые и тренировочные данные репрезентативно.

__3.__ __Что лучше 2 модели случайного леса по 500 деревьев или одна на 1000, при условии, что ВСЕ параметры кроме количества деревьев одинаковы?__

Прежде всего, случайный лес -- это алгоритм машинного обучения, заключающийся в использовании ансамбля решающих деревьев. По умолчанию голоса каждого дерева имеют одинаковый вес, а затем эти голоса усредняются. Предположим, что множества $X$ и $Y$ имеют одинаковый размер. В нашем случае, это голоса двух случайных лесов по 500 деревьев. Затем, если мы возьмём средние значения элементов множества $X$ и $Y$, а затем усредним эти два значения, то получим такой же результат, если бы мы изначально объединили множества $X$ и $Y$ ($Z = X + Y$) (в нашем случае, это голоса случайного леса, в котором 1000 деревьев), а после вычислили среднее значение элементов получившегося множества. Действительно,
$$
\overline{x} = \frac{1}{N} \sum_{n=1}^{N} x_{n}, \qquad
\overline{y} = \frac{1}{N} \sum_{n=1}^{N} y_{n},
$$
$$
\mathrm{res} = \frac{\overline{x} + \overline{y}}{2} = \frac{1}{2 N} \left( \sum_{n=1}^{N} x_{n} + \sum_{n=1}^{N} y_{n} \right) = \frac{1}{2 N}\sum_{n=1}^{2 N} z_{n}.
$$
Таким образом, заключаем, что эти две модели дают одинаковый результат. Все отклонения носят случайный характер и не являются статистически значимыми.

__Ответ__: модели дают одинаковый результат.

__5.__ __Есть выборка клиентов с заявкой на кредитный продукт. Датасет состоит из персональных данных: возраст, пол и т.д. Необходимо предсказывать доход клиента, который представляет собой непрерывные данные, но сделать это нужно используя только модель классификации.__

1. Проанализируем тренировочные данные и найдем самое высокое значение дохода клиентов Обозначим это значение как $Y_{\mathrm{high}}$.
2. На отрезке $[0, Y_{\mathrm{high}}]$ выберем $N$ различных точек, соответстующих доходу. Как именно выбирать эти точки является отдельной задачей. Логично выбирать точки так, чтобы расстояние между ними монотонно увеличивалось ближе к концу отрезка.
3. Будем тренировать $N$ модель бинарной классификации. Каждая модель будет выдавать ответ на вопрос больше ли доход клиента $Y_{i}$-ой точки ($i=1,2,\ldots,N$). Т.е. мы тренируем $N$ моделей на одних и тех же тренировочных данных.
5. Даём на вход каждой модели данные клиента для предсказания его дохода. Переводим результаты каждой из $N$ моделей в форму вероятности. Очевидно, что вероятность того, что доход больше или равен $0$ равна $1$ и близок к нулю, если доход больше или равен $Y_{N}$. В итоге, мы имеем $N+1$ точек со значениями вероятности.
6. Интерполируем полученные вероятности одним из известных методов (можно провести прямые линии между соседними точками, линейная интерполяция).
7. Выбираем подходящий для наших целей уровень значимости предсказаний наших моделей. Находим на графике величину дохода, которая соответствует выбранному уровню значимости.
